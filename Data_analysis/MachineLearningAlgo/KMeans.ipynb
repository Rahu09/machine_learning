```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
style.use("ggplot")
from sklearn.cluster import KMeans

# plotting and bisualizing out data before feeding it into Machine Learning Algorithm
x = [1, 5, 1.5, 8, 1, 9]
y = [2, 8, 1.8, 8, 0.6, 11]
plt.scatter(x,y)
plt.show()
```

```python
# Converting out data to a NumPy array
X = np.array([[1,2], [5,8], [1.5,1.8], [8,8], [1,0.6], [9,11]])

# We initialize K-means algo with the requierd parameter and we use .fit() to fit the data.
kmeans = KMeans(n_clusters=2,n_init=10)
kmeans.fit(X)

# Getting the values of centroids and labels based on the fitment
centroids = kmeans.cluster_centers_
labels = kmeans.labels_

print(centroids)
print(labels)
```

```python
# Plotting and visualizing output
colors = ["g.", "r.", "c.", "y."]
for i in range(len(X)):
    print("coordinate:", X[i], "label:", labels[i])
    plt.plot(X[i][0], X[i][1], colors[labels[i]], markersize = 10)

plt.scatter(centroids[:, 0], centroids[:, 1], marker = "x", s=150, linewidths=5, zorder=10)
plt.show()
```
